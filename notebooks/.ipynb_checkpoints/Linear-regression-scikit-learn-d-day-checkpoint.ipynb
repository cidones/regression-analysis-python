```python
import pandas                as pd 
import numpy                 as np
import matplotlib.pyplot     as plt
import seaborn               as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model    import LinearRegression
from sklearn                 import metrics

%matplotlib inline
```

## Simple Linear Regression 

```python
dataset = pd.read_csv('Weather.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
dataset.plot(x='MinTemp', y='MaxTemp', style='o')
plt.title('MinTemp vs MaxTemp')
plt.ylabel('MinTemp')
plt.xlabel('MaxTemp')
```

```python
#Plot the average MaxTemp
plt.figure(figsize=(15, 10))
plt.tight_layout()
sns.distplot(dataset['MaxTemp'])
```

```python
#Next step divide data into Attributes and Label
X = dataset['MinTemp'].values.reshape(-1, 1)
y = dataset['MaxTemp'].values.reshape(-1, 1)

#Split data 80% training and 20% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#Train the algorithm
regressor = LinearRegression()
regressor.fit(X_train, y_train) 


print("Intercept: ", regressor.intercept_) #Retrieve the intercept
print("Slope: ", regressor.coef_)   #Retrieve the slope
```

```python
#Predict
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

```python
#Plot
df1 = df.head(25)
df1.plot(kind='bar', figsize=(16, 10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

plt.show()
```

```python
#Plot straight line with test data
plt.scatter(X_test, y_test, color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

## Evaluate the performance of the algorithm

```python
print('Mean absolute error: ', metrics.mean_absolute_error(y_test, y_pred))
print('Mean squared error: ', metrics.mean_squared_error(y_test, y_pred))
print('Root squared error: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

## Multiple Linear Regression 

```python
newdataset = pd.read_csv('winequality.csv')
newdataset.head()
```

```python
newdataset.shape
```

```python
newdataset.describe()
```

```python
newdataset.isnull().any() # check if data contain NaN values
```

```python
newdataset = newdataset.fillna(method='ffill') #in case of True values remove all null values
```

```python
#Split data into X =  Attributes and  y = Labels
X = newdataset[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide','total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
y = newdataset['quality'] # if we include .values return a numpy array and not a pandas DataFrame
```

```python
plt.figure(figsize=(15,10)) # check the average of quality
plt.tight_layout()
sns.distplot(newdataset['quality'])
```

```python
#Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#Train model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Check coefficients the regression chose
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient']) 
#coeff_df = pd.DataFrame(regressor.coef_, columns=['Coefficient'])
coeff_df
```

```python
#Predict
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 = df.head(25)
df1
```
